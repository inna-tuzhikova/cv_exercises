```python
import numpy as np
import os
from keras.preprocessing.image import load_img, img_to_array
from keras.datasets import mnist
import matplotlib.pyplot as plt
```

```python
# PATH = '/home/inna/code/nn/hse_deeplearning/l4/hw/ds/'
# LABELS = ['female_kid', 'female_teenager', 'female_young_adult', 'female_middle_age', 'female_old',
#           'male_kid', 'male_teenager', 'male_young_adult', 'male_middle_age', 'male_old']

PATH = '/home/inna/code/cv/shapes_img_gen/tmp/'
LABELS = ['circle', 'square', 'triangle', 'pentagon']

HEIGHT = 512
WIDTH = 512
n_channels = 3
```

```python
def get_mean_img(img_dir):
    tmp = np.zeros(shape=(HEIGHT, WIDTH, n_channels))
    n_img = 0
    for f in os.listdir(img_dir):
        f_name = os.path.join(img_dir, f)
        if os.path.isdir(f_name):
            continue
        try:
            img = load_img(f_name, target_size=(HEIGHT, WIDTH))
        except Exception as e:
            print('Unable to load %s' % f_name)
            continue
        img = img_to_array(img)
        tmp += img
        n_img += 1
    return tmp / n_img
```

```python
mean_imgs = []
for i, label in enumerate(LABELS):
    mean_imgs.append(get_mean_img(os.path.join(PATH, label)))
```

```python
plt.figure(figsize=(10, 25))
for i, (label, img) in enumerate(zip(LABELS, mean_imgs)):
    plt.subplot(5, 2, i + 1)
    plt.imshow(img.astype(np.uint8))
    plt.title(label.replace('_', ' ').capitalize())
    plt.xticks([])
    plt.yticks([])
plt.tight_layout()
plt.show()
```

```python
(digits, labels), _ = mnist.load_data()
```

```python
mean_imgs = []
for i in range(10):
    idxs = [idx for idx in range(labels.shape[0]) if labels[idx] == i]
    mean_imgs.append(np.mean(digits[idxs], axis=0))
```

```python
plt.figure(figsize=(10, 4))
for i, (label, img) in enumerate(zip(labels, mean_imgs)):
    plt.subplot(2, 5, i + 1)
    plt.imshow(img.astype(np.uint8), cmap='gray')
    plt.xticks([])
    plt.yticks([])
plt.tight_layout()
plt.show()
```
