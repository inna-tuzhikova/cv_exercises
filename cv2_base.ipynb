```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('lena.png')
```

```python
def show_image_window(image):
    cv2.imshow('my_window', image)
    while True:
        if cv2.waitKey(0) & 0xFF == 27:
            break
    cv2.destroyAllWindows()
```

```python
def show_image_plt(image):
    plt.figure(figsize=(5, 5))
    if image.ndim == 2:
        plt.imshow(image, cmap='gray')
    else:
        plt.imshow(image[...,::-1])
    plt.xticks([])
    plt.yticks([])
    plt.show()
```

```python
show_image_window(img)
```

```python
show_image_plt(img)
```

```python
show_image_plt(img[:,::-1,:])
```

```python
show_image_plt(img[::-1,:,:])
```

```python
show_image_plt(img[::-1,::-1,:])
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
print(img_gray.shape)
show_image_plt(img_gray)
cv2.imwrite('lena_gray.png', img_gray)
```

```python
img_gray_rgb = cv2.cvtColor(img_gray, cv2.COLOR_GRAY2RGB)
print(img_gray_rgb.shape)
show_image_plt(img_gray_rgb)
cv2.imwrite('lena_gray_rgb.png', img_gray_rgb)
```

```python
tmp = cv2.imread('lena_gray.png')
print(tmp.shape)
tmp = cv2.imread('lena_gray.png', cv2.IMREAD_UNCHANGED)
print(tmp.shape)
tmp = cv2.imread('lena_gray_rgb.png')
print(tmp.shape)
tmp = cv2.imread('lena_gray_rgb.png',cv2.IMREAD_UNCHANGED)
print(tmp.shape)
```

```python
show_image_plt(img)
rgb = {0: 'Red', 1: 'Green', 2: 'Blue'}
for ch in [0, 1, 2]:
    print('%s channel:' % rgb[ch])
    # [..., ::-1] - BGR -> RGB
    show_image_plt(img[..., ::-1][..., ch])

print('Custom grayscale (1/3, 1/3, 1/3):')
show_image_plt((img[..., ::-1][..., 0] + img[..., ::-1][..., 1] + img[..., ::-1][..., 2]) / 3)

print('Custom grayscale (0.3, 0.59, 0.11):')
show_image_plt(0.3*img[..., ::-1][..., 0] + 0.59*img[..., ::-1][..., 1] + 0.11*img[..., ::-1][..., 2])
print('cv2.cvtColor')
show_image_plt(img_gray_rgb)
```

```python
lena = cv2.imread('lena.png')
lena_head = lena[200:380, 200:350, :]
show_image_plt(lena_head)
```

```python
tmp = np.hstack([lena_head, lena_head, lena_head])
tmp = np.vstack((tmp, tmp, tmp))
tmp = np.hstack([tmp, tmp[:,::-1,:]])
tmp = np.vstack([tmp, tmp[::-1,:,:]])
show_image_plt(tmp)
cv2.imwrite('lena_collage.png', tmp)
```
