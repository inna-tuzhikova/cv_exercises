```python
import cv2
from matplotlib import pyplot as plt
```

```python
img_1_path = 'i1.jpeg'
img_2_path = 'i2.jpeg'
```

```python
img_1 = cv2.imread(img_1_path)
img_2 = cv2.imread(img_2_path)
```

```python
def show_images_in_row(*imgs):
    n_imgs = len(imgs)
    
    plt.figure(figsize=(n_imgs*5, 5))
    for i, idx in enumerate(imgs):
        plt.subplot(1, len(imgs), i + 1)
        plt.imshow(imgs[i][...,::-1])
#         plt.title(LABELS[np.argmax(labels[idx])])
        plt.xticks([])
        plt.yticks([])
    plt.tight_layout()
```

```python
show_images_in_row(img_1, img_2, img_1-img_2)
```
